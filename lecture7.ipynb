```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from CSV files
sat_data = pd.read_csv('SAT.csv')
lunch_wait_time_data = pd.read_csv('lunch_wait_time.csv')
student_age_data = pd.read_csv('student_age.csv')

# Rename columns to be more descriptive
sat_data.columns = ['SAT']
lunch_wait_time_data.columns = ['Wait Time']
student_age_data.columns = ['Age']

# Generate histograms for each dataset
fig, axes = plt.subplots(3, 1, figsize=(8, 12))

# Histogram for SAT scores
axes[0].hist(sat_data['SAT'], bins=20, color='blue', edgecolor='black')
axes[0].set_title('Histogram of SAT Scores')
axes[0].set_xlabel('SAT Score')
axes[0].set_ylabel('Frequency')

# Histogram for Lunch Wait Times
axes[1].hist(lunch_wait_time_data['Wait Time'], bins=20, color='green', edgecolor='black')
axes[1].set_title('Histogram of Lunch Wait Times')
axes[1].set_xlabel('Wait Time (minutes)')
axes[1].set_ylabel('Frequency')

# Histogram for Student Ages
axes[2].hist(student_age_data['Age'], bins=20, color='orange', edgecolor='black')
axes[2].set_title('Histogram of Student Ages')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Calculate the average value for each dataset
average_sat = sat_data['SAT'].mean()
average_wait_time = lunch_wait_time_data['Wait Time'].mean()
average_age = student_age_data['Age'].mean()

# Calculate the probability that someone scores higher than 1300 on the SAT
prob_sat_above_1300 = (sat_data['SAT'] > 1300).mean()

# Calculate the probability that the wait time is more than 3 minutes
prob_wait_more_than_3 = (lunch_wait_time_data['Wait Time'] > 3).mean()

# Calculate the probability that someone is younger than the user
user_age = 21  # Assuming the user's age is 21
prob_younger_than_user = (student_age_data['Age'] < user_age).mean()

# Print the results
print(f"Average SAT Score: {average_sat}")
print(f"Average Wait Time (minutes): {average_wait_time}")
print(f"Average Age: {average_age}")
print(f"Probability SAT > 1300: {prob_sat_above_1300}")
print(f"Probability Wait Time > 3 minutes: {prob_wait_more_than_3}")
print(f"Probability Someone Younger than User (Age 21): {prob_younger_than_user}")
```

```python
from scipy.stats import norm

# Calculation 1: 1 - norm.cdf(67, 70, 3)
prob_greater_than_67 = 1 - norm.cdf(67, 70, 3)
print(f"Probability of value greater than 67 (mean=70, std=3): {prob_greater_than_67}")

# Calculation 2: norm.cdf(67, 64.5, 2.5)
prob_less_than_or_equal_67 = norm.cdf(67, 64.5, 2.5)
print(f"Probability of value less than or equal to 67 (mean=64.5, std=2.5): {prob_less_than_or_equal_67}")

# Calculation 3: norm.ppf(0.90, 64.5, 2.5)
percentile_90 = norm.ppf(0.90, 64.5, 2.5)
print(f"90th percentile (mean=64.5, std=2.5): {percentile_90}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mu_m = 70       
mu_f = 64.5     
sigma_m = 3    
sigma_f = 2.5  


x = np.linspace(55, 85, 1000)
x2 = np.linspace(52, 77, 1000)

pdf_m = stats.norm.pdf(x, mu_m, sigma_m)
pdf_f = stats.norm.pdf(x2, mu_f, sigma_f)
plt.figure(figsize=(10, 6))
plt.plot(x, pdf_m, color='blue', label='Male Gaussian PDF')
plt.plot(x2, pdf_f, color='red', label='Female Gaussian PDF')

x_fill = np.linspace(50, 67, 1000)
pdf_fill = stats.norm.pdf(x_fill, mu_f, sigma_f)
plt.fill_between(x_fill, pdf_fill, color='lightgrey', alpha=0.5, label='Shaded Region')

plt.xlabel('X')
plt.ylabel('Probability Density')
plt.title('Height distribution of male (blue) and female (red)')
plt.grid(True)

# Show plot
plt.show()
```
