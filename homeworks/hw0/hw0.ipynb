```python
#Question 2
import autograd.numpy as np
from scipy.integrate import quad

def f1(x):
    return 2*x**2 + 1

def f2(x):
    return np.exp(x)

def f3(x):
    return (3*x - 2)**2

def f4(x):
    return np.exp(-x) + 2*x + 1

auto_integral1, _ = quad(f1, 0, 1)
auto_integral2, _ = quad(f2, 0, 1)
auto_integral3, _ = quad(f3, 0, 1)
auto_integral4, _ = quad(f4, 0, 1)

theoretical_integral1 = 5 / 3
theoretical_integral2 = np.exp(1) - 1
theoretical_integral3 = 1
theoretical_integral4 = 3 - 1/np.exp(1)

print(f'1st Integral (Auto: {auto_integral1:.5f}, Theoretical: {theoretical_integral1:.5f})')
print(f'2nd Integral (Auto: {auto_integral2:.5f}, Theoretical: {theoretical_integral2:.5f})')
print(f'3rd Integral (Auto: {auto_integral3:.5f}, Theoretical: {theoretical_integral3:.5f})')
print(f'4th Integral (Auto: {auto_integral4:.5f}, Theoretical: {theoretical_integral4:.5f})')
```

```python
#Question 1
import autograd.numpy as np
from autograd import grad

def f1(x):
    return 2*x**2 + 1

def ᐁf1(x):
    return 4*x

def f2(x):
    return np.exp(x**2)

def ᐁf2(x):
    return 2*x*np.exp(x**2)

def f3(x):
    return (3*x - 2)**2

def ᐁf3(x):
    return 18*x - 12

def f4(x, y):
    return np.sum((3*x - y)**2)

def ᐁf4(x, y):
    return 6 * np.sum(3*x - y)

def f5(x):
    return np.log(1/x) / np.log(2)

def ᐁf5(x):
    return -1/(x * np.log(2))

def f6(x):
    return 2 * np.log(x - 2) / np.log(2)

def ᐁf6(x):
    return 2 / ((x - 2) * np.log(2))

auto_grad1 = grad(f1)
auto_grad2 = grad(f2)
auto_grad3 = grad(f3)
auto_grad5 = grad(f5)
auto_grad6 = grad(f6)

for i in range(5):
    x = np.random.randn() + 2
    y = np.array([1.0, 2.0, 3.0])

    print(f"1st Function (Auto: {auto_grad1(x):.5f}, Theoretical: {ᐁf1(x):.5f})")
    print(f"2nd Function (Auto: {auto_grad2(x):.5f}, Theoretical: {ᐁf2(x):.5f})")
    print(f"3rd Function (Auto: {auto_grad3(x):.5f}, Theoretical: {ᐁf3(x):.5f})")
    print(f"4th Function (Auto: {grad(lambda x: f4(x, y))(x):.5f}, Theoretical: {ᐁf4(x, y):.5f})")
    print(f"5th Function (Auto: {auto_grad5(x):.5f}, Theoretical: {ᐁf5(x):.5f})")
    print(f"6th Function (Auto: {auto_grad6(x):.5f}, Theoretical: {ᐁf6(x):.5f})")
    
```
