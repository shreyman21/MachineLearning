```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore

# Question 1
X = np.array([[0, 1], [1, 1]])
y = np.array([[0], [2]])
w = np.array([[2], [2]])
η = 0.1
n = X.shape[0]

# Minimize f(x) = 1/n * sum((w^T * xi - yi)^2)
# Gradient Descent
def f(w):
    return (1/n) * np.sum((X.dot(w) - y) ** 2)

def f_prime(w):
    return (2/n) * X.T.dot(X.dot(w) - y)

# Gradient Descent loop
f_value_list = []
for _ in range(20):
    w = w - η * f_prime(w)
    f_value_list.append(f(w))

print('Best w = \n', w)

plt.plot(f_value_list)
plt.title('MSE over GD')
plt.xlabel('steps')
plt.ylabel('MSE')
plt.show()
```

```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore

#Question 2
A = np.array([[0,1],[1,1]])
y = np.array([[0],[2]])
w = np.array([[2],[2]]) 
η = 0.1

def f(w):
	p = (np.dot(A,w) - y)
	return np.dot(np.transpose(p), p).item()

f_value_list = []
for i in range(100):
	fᑊ = 2*A.T.dot(A.dot(w) - y) 
	w = w - η*fᑊ			
	f_value_list.append(f(w))
 
print('Best w = \n', w)

plt.plot(f_value_list)
plt.title('MSE over GD')
plt.xlabel('steps')
plt.ylabel('MSE')
plt.show()
```
